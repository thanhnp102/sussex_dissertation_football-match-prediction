## Clean data & Extract features

### Football-data

```python
fd_df = fd_match[['Date','HomeTeam', 'AwayTeam','FTHG','FTAG','HS','AS','HST', 'AST','HF','AF','HC','AC','HY', 'AY', 'HR', 'AR']].copy(deep=True)
fd_df = fd_df.merge(dim_team[['team_name','fd_team']], left_on='HomeTeam', right_on='fd_team', how='left')
fd_df = fd_df.merge(dim_team[['team_name','fd_team']], left_on='AwayTeam', right_on='fd_team', how='left')

fd_df.rename(columns={
    'team_name_x':'home_team', 
    'team_name_y':'away_team',
    'FTHG':'home_goal',
    'FTAG':'away_goal',
    'HS':'home_shot',
    'AS':'away_shot',
    'HST': 'home_sot',
    'AST': 'away_sot',
    'HF': 'home_foul',
    'AF': 'away_foul',
    'HC': 'home_corner',
    'AC': 'away_corner',
    'HY': 'home_yellow',
    'AY': 'away_yellow',
    'HR': 'home_red',
    'AR': 'away_red'
}, inplace=True)

def parse_date(date_str):
    for fmt in ('%d/%m/%Y', '%d/%m/%y'):
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            pass

fd_df['match_date'] = fd_df['Date'].apply(parse_date)

fd_df = fd_df[['match_date', 'home_team', 'away_team','home_goal','away_goal','home_shot','away_shot','home_sot','away_sot','home_foul','away_foul','home_corner','away_corner','home_yellow','away_yellow','home_red','away_red']]

fd_df.head()
```

### Wyscout data

Extract performance metrics from event data

```python
# shots
# ws_team_shot = ws_event[(ws_event["eventName"] == "Shot") |
#                    (ws_event["subEventName"] == "Free kick shot")
#                 ].groupby(["matchId","teamId"]).size().reset_index(name='shot')
# ws_home_shot = ws_shot
# ws_home_shot = ws_home_shot.rename(columns={"matchId": "match_id","teamId": "home_team_id", "shot": "home_shot"})
# ws_away_shot = ws_shot
# ws_away_shot = ws_away_shot.rename(columns={"matchId": "match_id","teamId": "away_team_id", "shot": "away_shot"})

# passes
ws_pass = ws_event[(ws_event["eventName"] == "Pass")
                   ].groupby(["matchId","teamId"]).size().reset_index(name='pass')
ws_home_pass = ws_pass
ws_home_pass = ws_home_pass.rename(columns={"matchId": "match_id","teamId": "home_team_id", "pass": "home_pass"})
ws_away_pass = ws_pass
ws_away_pass = ws_away_pass.rename(columns={"matchId": "match_id","teamId": "away_team_id", "pass": "away_pass"})

# offsides
ws_offside = ws_event[(ws_event["eventName"] == "Offside")
                    ].groupby(["matchId","teamId"]).size().reset_index(name='offside')
ws_home_offside = ws_offside
ws_home_offside = ws_home_offside.rename(columns={"matchId": "match_id","teamId": "home_team_id", "offside": "home_offside"})
ws_away_offside = ws_offside
ws_away_offside = ws_away_offside.rename(columns={"matchId": "match_id","teamId": "away_team_id", "offside": "away_offside"})

# clearances
ws_clearance = ws_event[(ws_event["subEventName"] == "Clearance")
                        ].groupby(["matchId","teamId"]).size().reset_index(name='clearance')
ws_home_clearance = ws_clearance
ws_home_clearance = ws_home_clearance.rename(columns={"matchId": "match_id","teamId": "home_team_id", "clearance": "home_clearance"})
ws_away_clearance = ws_clearance
ws_away_clearance = ws_away_clearance.rename(columns={"matchId": "match_id","teamId": "away_team_id", "clearance": "away_clearance"})

# saves
ws_save = ws_event[(ws_event["eventName"] == "Save attempt")
                ].groupby(["matchId","teamId"]).size().reset_index(name='save')
ws_home_save = ws_save
ws_home_save = ws_home_save.rename(columns={"matchId": "match_id","teamId": "home_team_id", "save": "home_save"})
ws_away_save = ws_save
ws_away_save = ws_away_save.rename(columns={"matchId": "match_id","teamId": "away_team_id", "save": "away_save"})

# freekicks
ws_freekick = ws_event[(ws_event["eventName"] == "Free Kick") &
                        (ws_event["subEventName"].isin(["Free Kick", "Free Kick shot", "Free Kick cross"]))
                      ].groupby(["matchId","teamId"]).size().reset_index(name='freekick')
ws_home_freekick = ws_freekick
ws_home_freekick = ws_home_freekick.rename(columns={"matchId": "match_id","teamId": "home_team_id", "freekick": "home_freekick"})
ws_away_freekick = ws_freekick
ws_away_freekick = ws_away_freekick.rename(columns={"matchId": "match_id","teamId": "away_team_id", "freekick": "away_freekick"})
```

```python
ws_df = ws_match[['wyId', 'dateutc', 'label']].copy(deep=True)
ws_df['match_date'] = ws_df['dateutc'].apply(lambda x: x.split()[0])

# extract team ids
def extract_team_ids(teams_data):
    home_team_id = None
    away_team_id = None
    for team_id, team_info in teams_data.items():
        if team_info['side'] == 'home':
            home_team_id = team_info['teamId']
        elif team_info['side'] == 'away':
            away_team_id = team_info['teamId']
    return home_team_id, away_team_id
ws_df[['home_team_id', 'away_team_id']] = ws_match['teamsData'].apply(lambda x: pd.Series(extract_team_ids(x)))

ws_df['home_team'] = ws_df['home_team_id'].map(dim_team.set_index('ws_team_id')['team_name'])
ws_df['away_team'] = ws_df['away_team_id'].map(dim_team.set_index('ws_team_id')['team_name'])

ws_df['home_score'] = ws_df['label'].apply(lambda x: int(x.split(',')[1].split(' - ')[0]))
ws_df['away_score'] = ws_df['label'].apply(lambda x: int(x.split(',')[1].split(' - ')[1]))

ws_df.drop(columns=['label', 'dateutc'], inplace=True)
ws_df.rename(columns={'wyId':'match_id'}, inplace=True)

# ws_df = ws_df.merge(ws_home_shot, on=["match_id","home_team_id"], how="left")
# ws_df = ws_df.merge(ws_away_shot, on=["match_id","away_team_id"], how="left")
ws_df = ws_df.merge(ws_home_pass, on=["match_id","home_team_id"], how="left")
ws_df = ws_df.merge(ws_away_pass, on=["match_id","away_team_id"], how="left")
ws_df = ws_df.merge(ws_home_offside, on=["match_id","home_team_id"], how="left")
ws_df = ws_df.merge(ws_away_offside, on=["match_id","away_team_id"], how="left")
ws_df = ws_df.merge(ws_home_clearance, on=["match_id","home_team_id"], how="left")
ws_df = ws_df.merge(ws_away_clearance, on=["match_id","away_team_id"], how="left")
ws_df = ws_df.merge(ws_home_save, on=["match_id","home_team_id"], how="left")
ws_df = ws_df.merge(ws_away_save, on=["match_id","away_team_id"], how="left")
ws_df = ws_df.merge(ws_home_freekick, on=["match_id","home_team_id"], how="left")
ws_df = ws_df.merge(ws_away_freekick, on=["match_id","away_team_id"], how="left")



ws_df.head(10)
```

## Match result

```python
def match_result(row):
    if row['home_goal'] > row['away_goal']:
        return 'Home'
    elif row['home_goal'] < row['away_goal']:
        return 'Away'
    else:
        return 'Draw'
```

```python
df['result'] = df.apply(match_result, axis=1)

df.head()
```

```python
df = df[['match_date', 'home_team', 'away_team'] + metrics + ['result']]
df.head()
```
